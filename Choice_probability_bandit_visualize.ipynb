```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
plt.style.use('default')
std_list = [7,6,5,4]
TU = 8
n = len(std_list)

for (index,std) in enumerate(std_list):
    Data = pd.read_csv(f'saved/analysed_data_seek_init_std={std}_TU={TU}_RS_5_5_averse.csv')
    Data = Data[(Data.trial > 10) & (Data.trial <= 20)]
    X = []
    Y = []
    v = []
    for i in np.arange(-1.5, 1.5, 0.1):
        Data_H = Data.loc[abs(Data['Q1'] - Data['Q2'] - i) < 0.05]
        if len(Data_H) == 0:
            ratio = 0
        else:
            # Data_C = Data_C.reshape(Data_C.shape[0])
            ratio = len(Data_H[Data_H['choice'] == 0]) / len(Data_H)
        X.append(i)
        v.append(ratio)
    my_colors = cm.Greens(index/n+0.3)

    plt.plot(X, v, label=f'$\sigma_1^2={std},\sigma_2^2={TU-std} $',color=my_colors)

plt.axvline(0, 0, 1, linestyle='--', color='black')
plt.axhline(0.5, 0, 1, linestyle='--', color='black')
plt.xlabel('expected value difference',size=20)
plt.ylabel('choice probability',size=20)
plt.legend(prop={'size': 15})
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```
